```python
from collections import deque
import time
import csv
```

```python
### --------------------------------- NodeData Class --------------------------------------------### 

class NodeData(object):
    def __init__(self, i, xi,cb,cv):
        self.i = i          #level in tree and the index in "pt" table
        self.xi = xi        #number of units of prodect
        self.cb = cb        #actual profit in the current node
        self.cv = cv        #volume consumed in the current node
```

```python
### ------------------------------------- Node Class --------------------------------------------### 

class Node(object):
    def __init__(self, data, parent, indexInParentChilds, nextPostOrd=None, cycl=False):
        self.data = data
        self.parent = parent
        self.nextPostOrd = nextPostOrd
        self.indexInParentChilds = indexInParentChilds
        self.cycl = cycl
        self.childs = []
```

```python
### ---------------------to compute the smallest volume in product's table------------------------###
def minV_of_products(productsTable):
    if(len(productsTable)>0):
        minVolum = productsTable[0][1]
        for x in productsTable:
            if x[1] < minVolum:
                minVolum = x[1]
        return minVolum
```

```python
def ratio_b_over_v(xi):
    return xi[2]/xi[1]
```

```python
def load_pt_from_csvF():
    fil = open('knapsac_instance.csv')
    csvreader = csv.reader(fil)
    header=[]
    header=next(csvreader)
    rows = []
    i=1
    pt=[]
    l=[]
    for row in csvreader :
        l = ['x'+str(i), int(row[0]), int(row[1])]
        pt.append(l)
        i=i+1
    return pt
```

```python
### ---------------------------------- evaluation function ----------------------------------------###

# <vMax> & <pt> are global variables
def eval_funct(nodeData):
    minV = minV_of_products(pt)
    restVolm =vMax - nodeData.cv 
    if(restVolm < minV) or (nodeData.i >= len(pt)-1):
        return nodeData.cb #évaluation exacte qui correspond à une solution
    else:
        return nodeData.cb + ((pt[nodeData.i + 1][2] / pt[nodeData.i + 1][1])*(restVolm))
```

```python
def print_solution(solution,M):    
    print("optimal benefice == " + str(M))
    print("Les produits à mettre dans le  sac à dos sont : ")
    while (solution != None) and (solution.data.i !=-1):
        print("produit: " + str(pt[solution.data.i][0]) + " Xi = " + str(solution.data.xi))
        solution = solution.parent
```

```python
def Branch_Bound_without_stack(root,Vmax):
    current = root
    M = 0
    solution = root

    while current:
        
        rv = vMax - current.data.cv 
        npi = current.data.i +1

        if not current.cycl:
            if (npi<= len(pt)-1 ):
                
                nf = int( rv / pt[npi][1] )
                lowB=nf if (npi==len(pt)-1) else 0
                ipc=0
                
                for j in range(nf, lowB-1, -1):  
                    
                    #on calcule l'evaluation de noeud sans le créer 
                    fdata = NodeData(npi,j,cb=current.data.cb + j*(pt[npi][2]), cv = current.data.cv + j*(pt[npi][1]))      
                    evall = eval_funct(fdata)           
                    
                    #Si l'evaluation >= M+1 on crée le noeud
                    if(evall >= M+1):
                        f = Node(data=fdata, parent=current, indexInParentChilds=ipc)
                       
                        if(evall==fdata.cb):     #f est une solution 
                            M = evall
                            solution = f            
                        elif(f.data.i < len(pt)-1):     #f is not a leaf  #on devise 
                            current.childs.append(f)
                            ipc = ipc+1


        #Post Order Traversal managing
        l= len(current.childs)
        if l>0 and not current.cycl:
            current.childs[l-1].nextPostOrd = current
            current.cycl = True
            current = current.childs[0]
        else:
            if(current.parent):
                pl = len(current.parent.childs)
                if(current.indexInParentChilds == pl-1):
                    current = current.nextPostOrd
                else:
                    current = current.parent.childs[current.indexInParentChilds+1]
            else:
                current = None
                
    return M,solution
```

```python
### ----------------------------- test case 1 ------------------------------###


## instantiate products
x1 = ["x1",33,4]
x2 = ["x2",49,5]
x3 = ["x3",60,6]
x4 = ["x4",32,2]

vMax=130
    
pt = [x4,x2,x1,x3]
pt.sort(key=ratio_b_over_v, reverse=True) ## decreasing order of products by b/v ratio

    
#initialize tree root
rootData = NodeData(-1,-1,cb=0, cv=0)
treeRoot = Node(data = rootData, parent=None,indexInParentChilds=-1 )


start_time = time.time()
M,solution = Branch_Bound_without_stack(treeRoot,vMax)
ft = time.time() - start_time

print("\nsolution: \n")
print_solution(solution,M)
print("\n\n\n###---------- execution time == "+str(ft)+" ----------------- ###")
```
